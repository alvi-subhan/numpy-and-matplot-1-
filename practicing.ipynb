```python
import numpy as np
import datetime

days=["Thursday","Friday","Monday","Tuesday","Wednesday","Thursday","Friday","Monday","Tuesday","Wednesday","Thursday","Friday","Monday","Tuesday","Wednesday","Thursday","Friday","Monday","Tuesday","Wednesday"]
price=[109.9499969,112.8000031,114.0500031,111.9000015,110.4499969,112.3499985,113,115.25,117.5500031,122.5,122.1999969,122,120.5500031,121.1500015,120.8499985,120.5999985,122.0999985,121.8000031,120,124]
close=np.array(price)
volum=[48824400,25681200,13950300,10554000,7675400,8613400,8795700,16049700,16903900,19869100,15273700,14988100,4513100,5775500,5888400,4007500,5424900,2245500,3489800,2965400
]
volume=np.array(volum)
weekdays=["Monday","Tuesday","Wednesday","Thursday","Friday"]
day2num=[]
for day in days:
    if day=="Monday":
        day2num.append(0)
    elif day=="Tuesday":
        day2num.append(1)
    elif day=="Wednesday":
        day2num.append(2)
    elif day=="Thursday":
        day2num.append(3)
    elif day=="Friday":
        day2num.append(4)
day_in_num=np.array(day2num)
day_in_num
```

```python
days_avg=[]
for i in range(5):
    indices=np.where(day_in_num==i)
    prices=np.take(close,indices)
    avg=np.mean(prices)
    print("Day:",i,"\nprices:",prices,"\naverag",avg)
    days_avg.append(avg)
days_avg=np.array(days_avg) #converting the python array to numpy for applying functions
days_avg
```

```python
np.take(weekdays,np.argmax(days_avg)) #will return the day with the highest avg price
```

```python
returns=np.diff(np.log(close))
print(returns)

positive_return=np.where(returns>0)
print(positive_return)

good_days=np.take(days,positive_return)
print(good_days)
```

```python
max_vol=np.argmax(volume)

print(np.take(volume,max_vol)) #highest volume eversold

print(np.take(days,max_vol)) #this tells us when was the highest volume
```

```python
import numpy as np
dates, price, close=np.loadtxt('prices.csv', delimiter=',', usecols=(0,2,3), unpack=True, dtype='str')
#print(type(dates))
#print(type(price))
```

```python
i=0
str_date=[]
str_close=[]
str_price=[]
for date in dates:
    zz=date.split("'")
    str_date.append(zz[1])
for cl in close:
    clz=cl.split("'")
    str_close.append(clz[1])
for pr in price:
    prz=pr.split("'")
    str_close.append(prz[1])
#print(type(str_close[1]))   
```

```python
import datetime
def dates2num(s):
    return datetime.datetime.strptime(s, "%d-%m-%Y").date().weekday()
dates=list(map(dates2num,str_date))
dates=np.asanyarray(dates,'f')
price=np.asanyarray(str_price,'f')
close=np.asanyarray(str_close,'f')
```

```python
avgs=np.zeros(5)
for i in range(5):
    indices = np.where(dates == i)
    prices = np.take(close, indices)
    avg = np.mean(prices)
    print("Day", i, "\nprices", prices, "\nAverage", avg)
    avgs[i] = avg
    
print(avgs)
    
    
    
```
